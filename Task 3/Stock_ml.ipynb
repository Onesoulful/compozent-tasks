```python
le=LabelEncoder()
df['Name']= le.fit_transform(df['Name'])
df['Open']= le.fit_transform(df['Open'])
df['Percentage Change']= le.fit_transform(df['Percentage Change'])
df['Volumein Lakhs']= le.fit_transform(df['Volumein Lakhs'])
df['Turnover (crs.)']= le.fit_transform(df['Turnover (crs.)'])
df['365 d % chng']= le.fit_transform(df['365 d % chng'])
df['30 d % chng']= le.fit_transform(df['30 d % chng'])
df.head()
```

```python
x= df[['Open','Percentage Change','Volumein Lakhs','Turnover (crs.)','365 d % chng','30 d % chng']]#feature 
y=df['Name'] 
```

```python
x_train,x_test, y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
model= LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred =model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
